```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
import time

from matplotlib import animation, rc
from IPython.display import HTML

steps = 500

def Randomwalk1D(n):
    x = 0
    y = 0

    xposition = [0]
    yposition = [0] 

    for i in range (1,n+1):
        step = np.random.uniform(0,1)

        if step < 0.5:
            x += 1
            y += 1 

        if step > 0.5: 
            x += 1
            y += -1

        xposition.append(x)
        yposition.append(y)
    return [xposition,yposition]
   
Randwalk = Randomwalk1D(steps)


#fig, ax = plt.subplots()
#line, = ax.plot([], [], lw=2)


fig, ax = plt.subplots(1,2)
#fig = plt.figure()
#ax = fig.add_subplot(111)
line, = ax[1].plot([], [], lw=2)
fig.set_figwidth(9)
fig.set_figheight(3)

ax[0].plot(Randwalk[0],Randwalk[1], label = "Randwalk1D")
ax[0].set_title("Random Walk Output")
ax[1].set_title("Random Walk Step Animation")
plt.show()

def init():
    line.set_data([], [])
    return (line,)

x_data, y_data = [],[]

for i in range(steps):
    x_data.append(Randwalk[0][i])
    y_data.append(Randwalk[1][i])
    ax[1].set_xlim((0, i+5))
    ax[1].set_ylim(min(y_data)-5, (max(y_data)+5))
    line.set_data(x_data, y_data)
 
    fig.canvas.draw()
    fig.canvas.flush_events()
    time.sleep(0.01)

```
